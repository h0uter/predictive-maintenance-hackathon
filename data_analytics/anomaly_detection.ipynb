```python
import pickle

data = pickle.load(open('cleaned_data.pkl', 'rb'))
data.head()
```

```python
import banpei 
model   = banpei.SST(w=50)
results_A = model.detect(data["A"])
results_B = model.detect(data["B"])
results_C = model.detect(data["C"])

```

```python
import matplotlib.pyplot as plt

plt.plot(data["time"], results_A, label="A")
plt.plot(data["time"], results_B, label="B")
plt.plot(data["time"], results_C, label="C")
plt.legend()

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
data.plot(x="time", y=["A", "B", "C"], figsize=(20, 20))
```

```python
import numpy as np
tresh = 0.05
print((results_A > tresh).sum())
print((results_B > tresh).sum())
print((results_C > tresh).sum())

```
